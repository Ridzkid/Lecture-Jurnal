## Soal 2

Buatlah fungsi untuk menghitung jumlah kasus harian Covid-19 (dataset yang disediakan merupakan jumlah kasus kumulatif)!

```python
# Kumulatif= penjumlahan dari data sebelumnya. Untuk menghitung kasus harian, 
#cukup mengurangi nilai saat ini dengan sebelumnya

#Sebelum menghitung kasus harian, tanggal diubah menjadi baris dan negara menjadi Kolom

def daily_case(dat):
    dat=dat.set_index('Country/Region')
    dat=dat.drop(columns=['Province/State','Lat','Long']).transpose()
    
    return dat.diff().fillna(0)
```

```python
daily=daily_case(data)
daily
```

## Soal 4

Buatlah sebuah list dimana berisikan jumlah kasus harian Covid-19 dengan ketentuan hanya mengambil data Indonesia saja dan jumlah kasus di bawah 10000. Berapa jumlah hari dengan kasus di bawah 10000?

```python
indo=[]

for case in daily['Indonesia']:
    if case<10000:
        indo.append(case)

print(len(indo))
```
